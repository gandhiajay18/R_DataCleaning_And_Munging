### XML to Data Frame

#### Install and load XML Library and plyr library

```r
#install.packages("XML", repos='http://cran.us./r-project.org')
library(XML) 
library(plyr)
```

#### Take XML data Link, parse it to data, and convert it into data frame object

```r
datalink <- "http://aiweb.cs.washington.edu/research/projects/xmltk/xmldata/data/courses/reed.xml"
xml_data <- xmlParse(datalink, useInternalNodes = TRUE)         #Parse xml data from the link
#xml_df <- xmlToDataFrame(xml_data)
xml_list <- xmlToList(xml_data)             #Convert data to list
```

#### Convert Child node to Sibling nodes of the Parent - i.e. Turn into Columns

```r
#Ldply and unlist used to reiterate over all child nodes of XML and to create them as seperate columns in the data frame.
xml_df <- ldply(xml_list, function(x){ data.frame(t(unlist(x)))},.id=NULL)
```

#### Check the dataframe

```r
list(xml_df)
```

### Data Frame to CSV

#### Write the Dataframe into CSV

```r
write.csv(xml_df,file="xml_2_CSV.csv") #Create a CSV file in the given directory with the specified name "xml_2_CSV.csv"
```

#### Check if properly written

```r
xml2csv <- read.csv("xml_2_CSV.csv") #Read again to see if the CSV is actually created
list(xml2csv) #To verify
```
